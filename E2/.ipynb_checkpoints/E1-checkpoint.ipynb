### Importamos las librerias necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Cargamos nuestro archivo csv para poder graficar

```python
tablaCasos = "CasosNuevosConSintomas.csv"
df = pd.read_csv(tablaCasos)
df
```

### Realizamos la 1era gráfica

```python
df['TotalCasosPorRegion'] = df.sum(axis=1)
valores1 = df[["Region","2021-09-08"]]
valores2 = df[["Region","TotalCasosPorRegion"]]
ax = valores1.plot.bar(x="Region", y="2021-09-08", rot=90, title="Casos por región en el tiempo", color="purple")
ax = valores2.plot.bar(x="Region", y="TotalCasosPorRegion", rot=90, title="Casos por región en el tiempo acumulado", color="purple")
```
