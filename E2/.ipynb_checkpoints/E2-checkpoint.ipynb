### Importamos las librerias necesarias y definivmos los párametros para las gráficas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
```

### Cargamos nuestro archivo csv para poder graficar

```python
tablaCasos = "CasosNuevosConSintomas.csv"
df = pd.read_csv(tablaCasos)
df
```

### Realizamos la 1era gráfica

```python
df['TotalCasosPorRegion'] = df.sum(axis=1)
valores1 = df[["Region","2021-09-08"]]
valores2 = df[["Region","TotalCasosPorRegion"]]
ax = valores1.plot.bar(x="Region", y="2021-09-08", rot=90, title="Casos por región en el tiempo", color="purple")
ax = valores2.plot.bar(x="Region", y="TotalCasosPorRegion", rot=90, title="Casos por región en el tiempo acumulado", color="purple")
```

### Agrupamos los datos necesarios para la 2da gráfica

```python
MZnorte = df.iloc[0] + df.iloc[1] + df.iloc[2] + df.iloc[3]
MZcentro = df.iloc[4] + df.iloc[5]
Metropolitana = df.iloc[6]
MZcsur = df.iloc[7] + df.iloc[8] + df.iloc[9] +df.iloc[10]
MZsur = df.iloc[11] + df.iloc[12] + df.iloc[13]
MZaustral = df.iloc[14] + df.iloc[15] + df.iloc[16]
```
